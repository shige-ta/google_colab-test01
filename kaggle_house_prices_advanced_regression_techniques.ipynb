```python
import pandas as pd
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
sample_submission = pd.read_csv('sample_submission.csv')
```

```python
from pycaret.classification import *
from pycaret.regression import * 
from pycaret.utils import enable_colab
enable_colab()

exp_house_prices = setup(data = train, target = 'SalePrice')
```

```python
compare_models()
```

```python
ridge = create_model('ridge')
```

```python
ridge_tune = tune_model(ridge)
```

```python
ridge_predict = predict_model(ridge_tune)
```

```python
plot_model(ridge_tune, plot='feature')
```

```python
final_ridge_tune = finalize_model(ridge_tune)
result = predict_model(final_ridge_tune, data = test)
result
```

```python
sub = pd.DataFrame(pd.read_csv('test.csv')['Id'])
sub['SalePrice'] = list(map(int, result['Label']))
sub.to_csv('submission.csv', index=False)
```
