```python
import pandas as pd
train = pd.read_csv('train.csv', nrows = 10000)
# train = pd.concat((r for r in train), ignore_index=True) 
test = pd.read_csv('test.csv')
```

```python
from pycaret.classification import *
from pycaret.regression import * 
from pycaret.utils import enable_colab
enable_colab()

exp = setup(data = train, target = 'fare_amount')
```

```python
compare_models()
```

```python
ridge = create_model('ridge')
```

```python
ridge_tune = tune_model(ridge)
```

```python
ridge_predict = predict_model(ridge_tune)
```

```python
final_ridge_tune = finalize_model(ridge_tune)
result = predict_model(final_ridge_tune, data = test)
result
```

```python
sub = pd.DataFrame(test['key'])
sub['fare_amount'] = list(map(int, result['Label']))
sub.to_csv('submission.csv', index=False)
```
